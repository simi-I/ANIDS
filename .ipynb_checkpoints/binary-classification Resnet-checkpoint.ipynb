```python
def convert2image(row): 
    img = row.reshape(6,13)  # dimensions of the image
    resized = cv2.resize(img, (32,32))
    image = np.zeros((32,32,3))  # empty matrix
    image[:,:,0] = resized
    image[:,:,1] = resized
    image[:,:,2] = resized
    return image
```

```python
y_train_sampled = np.select([y_train_sampled == 2, y_train_sampled == 3, y_train_sampled == 4, y_train_sampled == 5, y_train_sampled == 6], [1,1,1,1,1], y_train_sampled)

y_val = np.select([y_val == 2, y_val == 3, y_val == 4, y_val == 5, y_val == 6], [1,1,1,1,1], y_val)

y_test = np.select([y_test == 2, y_test == 3, y_test == 4, y_test == 5, y_test == 6], [1,1,1,1,1], y_test)
```

```python
X_train_array = []

for i in range(X_train_sampled.shape[0]):  #data.shape[0] gives no. of rows
    face = X_train_sampled[i]  # remove one row from the data
    img = convert2image(face)  # send this row of to the function 
    X_train_array.append(img)
    
```

```python
X_val_array = []

for i in range(X_val.shape[0]):  #data.shape[0] gives no. of rows
    face = X_val[i]  # remove one row from the data
    img = convert2image(face)  # send this row of to the function 
    X_val_array.append(img)
    
```

```python
X_test_array = []

for i in range(X_test.shape[0]):  #data.shape[0] gives no. of rows
    face = X_test[i]  # remove one row from the data
    img = convert2image(face)  # send this row of to the function 
    X_test_array.append(img)
```

```python
X_train_array = np.array(X_train_array)
print(X_train_array.shape)

X_test_array = np.array(X_test_array)
print(X_test_array.shape)

X_val_array = np.array(X_val_array)
print(X_val_array.shape)
```

```python
### One hot encoding
onehot_encoder = OneHotEncoder(sparse=False)
```

```python
integer_encoded = y_train_sampled.reshape(len(y_train_sampled), 1)

y_train_encoded = onehot_encoder.fit_transform(integer_encoded)

integer_encoded = y_val.reshape(len(y_val), 1)

y_val_encoded = onehot_encoder.fit_transform(integer_encoded)

integer_encoded = y_test.reshape(len(y_test), 1)

y_test_encoded = onehot_encoder.fit_transform(integer_encoded)
```

```python
image_size = 32
batch_size = 32
input_shape = (image_size, image_size, 3)
# Create the base model from the pre-trained model VGG19
base_model = ResNet50(input_shape=input_shape, include_top=False, weights='imagenet')

# Let's take a look at the base model architecture
base_model.summary()
```

```python
# freeze the model weights
base_model.trainable = False
base_model.summary()
```

```python
# construct the head of the model that will be placed on top.....
head_model = base_model.output
head_model = Flatten(name="flatten")(head_model)

head_model = Dense(512, activation="relu")(head_model)
head_model = Dense(256, activation="relu")(head_model)
head_model = Dense(64, activation="relu")(head_model)
head_model = Dense(32, activation="relu")(head_model)
head_model = Dense(2, activation='softmax')(head_model)
model = Model(inputs=base_model.input, outputs=head_model)

model.summary()
```

```python
# Compiling the model
base_learning_rate = 0.0001

model.compile(Adam(learning_rate=base_learning_rate, decay=base_learning_rate/40), loss='categorical_crossentropy', metrics=['accuracy'])

earlystopping = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=25, restore_best_weights=True)

# save the model with least validation loss
filepath="resnet50_bestB.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)

# callback_list = [checkpoint, earlystopping, learning_rate_reduce]
callback_list = [checkpoint,earlystopping]
```

```python
initial_epochs = 30

tic = time.perf_counter()

history = model.fit(X_train_array, y_train_encoded,
                    batch_size = 128, 
                    epochs = initial_epochs,
                    validation_data = (X_val_array, y_val_encoded),
                    callbacks = callback_list,
                    verbose = 1)

toc = time.perf_counter()
print(f"Completed in {toc - tic:0.4f} seconds")
```

```python
model.evaluate(X_test_array, y_test_encoded)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python


# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(2):
    fpr[i], tpr[i], _ = roc_curve(y_test_encoded[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_encoded.ravel(), y_pred.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])


# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_encoded.ravel(), y_pred.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
# First aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(2)]))

# Then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for i in range(2):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

# Finally average it and compute AUC
mean_tpr /= 7

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
```
